# Create Train/Test for 7 Percentages <em>then stop</em>

```python
df_all = pd.read_csv("../data_balanced.csv")
df_all
```

```python
df_train = pd.DataFrame() # define empty dataframe
df_test = pd.DataFrame()

df_ = df_all.drop(['Unnamed: 0','level_0','index'], axis=1)#.drop_duplicates()  # drop duplicates
df  = shuffle(df_)

print("shape after removing duplicate:",df.shape)
df
```

```python
fraction = 0.9  # percent data
# 0.01, 0.05, 0.1, 0.3, 0.5, 0.7, 0.9

label_class = set(df['type_2'].values)
print("lables are:",label_class)


for i in label_class: # for each label in dataset
    df1 = df[df['type_2']==i]  # seprate each data points from each class
    print("total samples in label class\t",i,":\t",df1.shape)
    df2 = df1.sample(frac = fraction)
    print(" training samples in label class",i,":\t", df2.shape)
    df_train = df_train.append(df2,ignore_index = True) # append dataframe to training
    
    df4 = df1.drop(df2.index)
    print(" testing samples in label class\t",i, ":\t",df4.shape)
    df_test = df_test.append(df4, ignore_index = True) # append dataframe to test
```

```python
print("shape of traning data is",df_train.shape)
print("shape of test data is",df_test.shape)
```

```python
df_train.to_csv("./new_data_train_%s.csv"%(fraction),index=False)
df_test.to_csv("./new_data_test_%s.csv"%(fraction),index=False)
```

```python
a = []

k_values = np.arange(1,20,1)
fractions = [0.01, 0.05, 0.1, 0.3, 0.5, 0.7, 0.9]

for fraction in fractions:
    df_train = pd.read_csv("./new_data_train_%s.csv"%(fraction))
    df_test = pd.read_csv("./new_data_test_%s.csv"%(fraction))

    
    X = df_train.iloc[:,0:5].values # features
    y = df_train.iloc[:,-1].values.ravel() # class or label

    # df_test = shuffle(df_test)
    X_test  = df_test.iloc[:,0:5].values
    y_test  = df_test.iloc[:, -1].values.ravel()
    a = []
    for k in k_values: 
        classifier = KNeighborsClassifier(metric="manhattan", n_neighbors=k)
        scores = cross_val_score(classifier, X, y, cv=10)
        print((scores.mean(), scores.std() * 2))
        
        
        a.append("%0.5f" % (scores.mean()))
    optimal_k = a.index(max(a)) + 1
    print('The optimal value of k for', fraction, '% is:', optimal_k)
    print("Optimal KNN KFold Train Accuracy for",fraction,"% is:", max(a))
    
    clf = KNeighborsClassifier(n_neighbors=optimal_k)
    clf.fit(X,y)
    
    # test
    y_pred = clf.predict(X_test)
    test_acc = accuracy_score(y_test,y_pred)
    print("Optimal KNN KFold Test Accuracy for",fraction,"% is:",test_acc)
```

```python
k_values = np.arange(1,20,1)
fractions = [0.01, 0.05, 0.1, 0.3, 0.5, 0.7, 0.9]

for fraction in fractions:
    df_train = pd.read_csv("./new_data_train_%s.csv"%(fraction))
    df_test = pd.read_csv("./new_data_test_%s.csv"%(fraction))

    
    X = df_train.iloc[:,0:5].values # features
    y = df_train.iloc[:,-1].values.ravel() # class or label

    # df_test = shuffle(df_test)
    X_test  = df_test.iloc[:,0:5].values
    y_test  = df_test.iloc[:, -1].values.ravel()
    cv_scores = []
    for k in k_values: 
        classifier = KNeighborsClassifier(metric="manhattan", n_neighbors=k)
        scores = cross_val_score(classifier, X, y, cv=10)
        cv_scores.append("%0.4f"%(scores.mean()))
    
    for i in range(0,len(cv_scores)):
        cv_scores[i] = float(cv_scores[i])

    mse = [1-x for x in cv_scores]
    optimal_k = k_values[mse.index(min(mse))] 
        
        
    print('The optimal value of k for', fraction, '% is:', optimal_k)
    print("Optimal KNN KFold Train Accuracy for",fraction,"% is:", (1-min(mse))*100)
    
    clf = KNeighborsClassifier(n_neighbors=optimal_k)
    clf.fit(X,y)
    
    # test
    y_pred = clf.predict(X_test)
    test_acc = accuracy_score(y_test,y_pred)
    print("Optimal KNN KFold Test Accuracy for",fraction,"% is:",test_acc)
```
